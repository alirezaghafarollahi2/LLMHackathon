# Example: Two-agent conversation between a scientist and an engineer

```python
import os

from autogen import ConversableAgent, LLMConfig
llm_config=LLMConfig(api_type="openai", model="gpt-4o-mini")

with llm_config:
    scientist_agent = ConversableAgent(
        name="scientist",
        system_message="You are a sophisticated scientist. You response through the lens of science. Provide short but accurate response.",
    )
    engineer_agent = ConversableAgent(
        name="engineer_agent",
        system_message="You are an advanced engineer. You proide feedback about ideas and provide constructive feedback. Provide short but accurate response.",
    )

chat_result = engineer_agent.initiate_chat(
    scientist_agent,
    message="How can we address sustainability with new materials?",
    summary_method="reflection_with_llm",
    summary_args={"summary_prompt": "Provide a summary of the conversation. List the key points and takeaways."},
    max_turns=10, #max_turns (int or None): the maximum number of turns for the chat between the two agents. One turn means one conversation round trip. 
)
```

```python
print(chat_result.summary)
```
